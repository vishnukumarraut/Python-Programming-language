## **While loops**

```python
n=int(input("Enter the number :"))
i=1
while i<11:
  print(n,'*',i,'=',n*i)
  i+=1
```

```python
n=int(input("Enter the number :"))
i=1
while i<11:
  print(n*i)
  i+=1
```

#### while loop with else

```python
x=1

while x<3:
  print(x)
  x+=1
else:
  print("limit crossed!")
```

```python
# Gassing game
import random

jackpot = random.randint(1, 100)

while user!=jackpot:
    user = int(input("Guess kro: "))

    if user < jackpot:
        print("Galat! Higher guess kro.")
    elif user > jackpot:
        print("Galat! Lower guess kro.")
    else:
        print("Correct guess!")
        break
```

## Nested loop

```python
for i in range(1, 5):
    for j in range(1, 5):
        print(f"({i}, {j})")
```

```python
n=int(input("Enter a Number : "))
for i in range(1,n+1):
  for j in range(1,i+1):
    print("*",end="")

  print("")


```

```python
n=int(input("Enter a Number : "))
for i in range(1,n+1):
  for j in range(1,i+1):
    print(j,end="")

  print("")


```

```python
n=int(input("Enter a Number : "))
for i in range(1,n+1):
  for j in range(1,i+1):
    print(j,end="")
  for k in range(i-1,0,-1):
    print(k,end="")
  print("")


```

## Loop Control Statements

### • Break Statements

```python
for i in range(1,10):
  if i==9:
    break
  print(i)
```

```python
n=int(input("Enter a number : "))
for i in range(2,n):
  if(n%i==0):
    print("No is not prime")
    break
else:
  print("No is prime")
```

```python
from io import UnsupportedOperation
lower=int(input("Enter a lower range :"))
upper=int(input("Enter a upper range :"))
for i in range(lower,upper+1):
  for j in range(2,i):
    if i%j==0:
      print(f"the {i} is not a prime no.")
      break
  else:
    print(f"No {i} is Prime")

```

• Continue Statements

```python
for i in range(1,10):
  if i==9:
    continue
  print(i)
```
