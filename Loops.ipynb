# **Loops in python**


## **While loops**

```python
n=int(input("Enter the number :"))
i=1
while i<11:
  print(n,'*',i,'=',n*i)
  i+=1
```

```python
n=int(input("Enter the number :"))
i=1
while i<11:
  print(n*i)
  i+=1
```

#### while loop with else

```python
x=1

while x<3:
  print(x)
  x+=1
else:
  print("limit crossed!")
```

```python
# Gassing game
import random

jackpot = random.randint(1, 100)

while user!=jackpot:
    user = int(input("Guess kro: "))

    if user < jackpot:
        print("Galat! Higher guess kro.")
    elif user > jackpot:
        print("Galat! Lower guess kro.")
    else:
        print("Correct guess!")
        break
```
