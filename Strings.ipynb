```python
s = input("Enter a palindrome: ")
flag = True

for i in range(0, len(s) // 2):
    if s[i] != s[len(s) - i - 1]:
        flag = False
        print("Not a palindrome")
        break

if flag:
    print("Palindrome")

```

```python
digits = '0123456789'

number = int(input('Enter the number: '))
result = ''

while number != 0:
    result = digits[number % 10] + result
    number = number // 10

print(result)
print(type(result))
```
